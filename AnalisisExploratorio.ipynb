```python
# Uso replace_columns para reemplazarlas por los nuevos nombres

def replace_columns(df, nuevas_columnas):
  nuevo_nombre_columnas = {
    nombre_original: category + '_' + nuevo_nombre # Para agregar en el nombre el prefijo de la categoría, por ejemplo 'perfil', 'trabajo'.
    for category, cols in nuevas_columnas.items()
    for nombre_original, nuevo_nombre in cols.items()
  }
  return df.rename(columns=nuevo_nombre_columnas)
```

```python
# Creo el nuevo dataframe con columnas renombradas y lo visualizo
df_renamed = replace_columns(df, nuevas_columnas)
df_renamed.head()
```

```python
# Veo la info del dataframe
df_renamed.info()
```

## Analisis de Salario


```python
# Creamos un df nuevo sólo con las columnas que nos interesan
df_salario = df_renamed [["perfil_genero", "perfil_nivel_estudios", "salario_mensual_BRUTO"]]
```

```python
print (df_salario)
```

```python
# Calculamos sueldo promedio segun nivel de estudio
df_salario.groupby(by="perfil_nivel_estudios").mean()
```

```python
#Calculamos sueldo segun el genero
df_salario.groupby (["perfil_genero"]).mean()
```

## Analisis tipo Matriz


```python
# creo un nuevo df solo con las columnas que me interesan
df_matriz= df_renamed[["perfil_genero", "perfil_nivel_estudios"]]
```

```python
# agrupo por categoria y uso el "size" para conocer el tamaño de cada grupo
df_matriz.groupby (["perfil_nivel_estudios","perfil_genero"]).size()
```

```python
# Otra opcion es usar el counts
df_matriz.value_counts().sort_index()
```

```python
# Identificar el grupo que más gana
df_salario.groupby(["perfil_genero","perfil_nivel_estudios"]).mean().sort_index()
```

```python
# otra opcion para visualizar agrupado y ordenado
df_salario.groupby(by=["perfil_genero", "perfil_nivel_estudios"]).mean().sort_values(by="salario_mensual_BRUTO")
```

```python
# Creamos uno df nuevo para analizar salario neto por provincia
df_salario_Neto = df_renamed[["perfil_genero", "trabajo_provincia", "salario_mensual_NETO"]]

df_salario_Neto.groupby(by="trabajo_provincia").mean().sort_values(by="salario_mensual_NETO")
```

```python
df_salario_Neto.groupby(by="trabajo_provincia").mean().sort_values(by="salario_mensual_NETO", ascending = False)
```
